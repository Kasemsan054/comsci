<h5>
การเดินทางด้วยเรือสำราญ (Cruise ship traveling) เป็นธุรกิจขนาดใหญ่ ซึ่งสายการล่องเรือหลายสายกำลังแข่งขันการในลูกค้าทุกกลุ่มอายุ ทั้งในระดับเศรษฐกิจ และสังคม บริษัทนำเสนอการล่องเรือทุกประเภท ไม่ว่าจะทริปล่องเรือ 3 – 4 วันราคาไม่แพงในทะเลแคริบเบียน ถึงทริปล่องเรือ 12 ถึง 15 วันในทะเลเมดิเตอร์เรเนียน และรวมถึงทริปล่องเรือรอบโลกหลายเดือน ซึ่งธุรกิจเรือสำราญมีคุณสมบัติหลายอย่างที่ดึงดูดลูกค้าได้มากมาย เช่น ลูกค้าสามารถจองล่วงหน้าหกเดือนขึ้นไป โดยทริปล่องเรือ บริษัทจะนำเสนอบริการที่หลากหลาย ไม่ว่าจะเป็นการบริการด้านอาหาร ทริปท่องเที่ยวตามท่าเรือจุดแวะพัก เพื่อให้ลูกค้าได้ประสบการณ์ และรวมถึงการบริการดนตรี และความบันเทิงที่หลากหลายในช่วงเวลาเย็นและค่ำ <br>
ซึ่งการบริการความบันเทิงที่หลากหลาย เป็นสิ่งที่ท้าทายของธรุกิจการล่องเรือ เนื่องจากทริปล่องเรือโดยทั่วไปอาจมีผู้โดยสารมากกว่า 1,000 คน รวมทั้งคนโสดและคนมีคู่ ผู้สูงอายุ วัยกลางคน ผู้โดยสารที่มีหรือไม่มีบุตร และหนุ่มสาวที่มาฮันนีมูน ซึ่งผู้โดยสารประเภทต่างๆ เหล่านี้มีรสนิยมที่แตกต่างกันในด้านการบริการด้านความบันเทิงยามหลังอาหารค่ำ ซึ่งบางคนอาจต้องการเต้นรำแบบดั้งเดิม ดนตรีที่แตกต่างบางคนต้องการชมตลก บางคนต้องการดนตรีร็อค บางคนต้องการดูภาพยนตร์ บางคนต้องการพักผ่อนอ่านหนังสือ และอื่นๆ <br>
ดังนั้นการจัดการการบริการด้านความบันเทิงบนเรือสำราญ จึงต้องการมอบความหลากหลายของความบันเทิงที่ลูกค้าต้องการ ภายในงบประมาณที่เหมาะสม เนื่องจากหากลูกค้าพึงพอใจ ลูกค้าก็มักจะเป็นลูกค้าประจำ ซึ่งคำถามคือทำอย่างไร ในทริปการล่องเรือจึงสามารถจัดสรรความบันเทิงให้กับลูกค้าได้อย่างลงตัว <br>
เช่น การแสดงวงดนตรีที่เล่นเพลงเต้นรำทุกคืนในเลานจ์ที่ใหญ่ที่สุดของเรือสำราญ ซึ่งมีนักเปียโนเล่นทุกคืน นักเต้นเล่นการแสดงประเภทบรอดเวย์ รวมถึงรักร้องและนักแสดงตลกมืออาชีพ ซึ่งแม้ว่าความบันเทิงที่นำเสนอไม่มีค่าใช้จ่ายสำหรับลูกค้าทุกคน แต่ส่วนใหญ่มีลูกค้าเข้าร่วมน้อย หรือบางครั้งไม่มีเลย ซึ่งบางครั้งมีการแสดงของนักร้องมืออาชีพ นักเต้นและนักแสดงตลกที่สามารถดึงดูดผู้ชมจำนวนมาก แต่ก็ยังมีที่นั่งว่างอยู่มากมาย ทั้งๆที่พนักงานล่องเรือได้โพสต์ตารางประจำสัปดาห์แล้ว เป็นต้น<br>

#### **หากคุณเป็นผู้รับผิดชอบด้านการจัดสรรค์ความบันเทิงของทริปเรือสำราญ คุณจะอธิบายปัญหาที่เกิดขึ้นอย่างไร ระหว่างความบันเทิงที่นำเสนอและความบันเทิงที่ลูกค้าต้องการ? และจะพยายามแก้ปัญหาอย่างไร**

**ปัญหาที่เกิดขึ้น:**

1. ขาดความเข้าใจความต้องการของลูกค้า: การแสดงความบันเทิงที่เกิดขึ้นบนเรือสำราญอาจมาจากการตัดสินใจของผู้รับผิดชอบด้านการจัดสรรความบันเทิง โดยอิงจากความคาดหวังหรือความถนัดของทีมงานภายใน ซึ่งอาจไม่ตรงกับความสนใจหรือความต้องการของลูกค้าที่มาใช้บริการ
2. การตั้งตารางเวลาการบริการที่ไม่เหมาะสม: การพิจารณาจัดสรรเวลาในการแสดงความบันเทิงบนเรือสำราญอาจไม่ถูกต้อง อาจจะมีการให้บริการแสดงความบันเทิงที่น่าสนใจมากเกินไปในวันหนึ่ง และในวันอื่นๆ อาจมีการแสดงที่น้อยลง ทำให้ลูกค้าบางกลุ่มรู้สึกไม่พอใจหรือไม่ต้องการมาเข้าร่วม
3. ขาดการตรวจสอบความพอใจของลูกค้า: อาจไม่มีการตรวจสอบหรือสำรวจความพอใจของลูกค้าต่อความบันเทิงที่ได้รับในทริปการล่องเรือ ทำให้ไม่ทราบถึงความพอใจและความไม่พอใจของลูกค้า ซึ่งทำให้ไม่สามารถปรับปรุงหรือแก้ไขปัญหาที่เกิดขึ้นได้

**แก้ปัญหา:**

1. ฟังเสียงของลูกค้า: ให้ความสำคัญในการฟังเสียงและความต้องการของลูกค้า สามารถใช้การสำรวจความพึงพอใจหลังจากการแสดงความบันเทิงหรือการให้บริการเสมอ นอกจากนี้ยังสามารถให้ลูกค้ามีช่องทางในการแสดงความคิดเห็นและแนะนำความต้องการของพวกเขา
2. ปรับปรุงตารางการแสดงความบันเทิง: การวางแผนตารางการแสดงความบันเทิงให้เหมาะสมกับความต้องการของลูกค้า โดยสามารถเปลี่ยนแปลงการแสดงในแต่ละวันหรือช่วงเวลา ให้เพิ่มความหลากหลายและน่าสนใจเพื่อให้ทุกคนสามารถเพลิดเพลินและเข้าร่วมความบันเทิงที่ตนเองต้องการได้
3. สร้างกลุ่มเป้าหมายของลูกค้า: จัดกลุ่มเป้าหมายของลูกค้าให้เหมาะสมกับความบันเทิงที่นำเสนอ เช่น กลุ่มที่ชอบดนตรีป๊อปอาจต้องการการแสดงดนตรีป๊อปเป็นส่วนใหญ่ ในขณะที่กลุ่มที่ชอบการเต้นรำอาจต้องการการแสดงเต้นรำมากขึ้น
4. ทดสอบและปรับปรุง: ทดสอบและปรับปรุงระบบการจัดสรรความบันเทิงอย่างต่อเนื่อง โดยใช้ข้อมูลและความคิดเห็นจากลูกค้า ทำการปรับปรุงเพื่อให้การบริการตรงกับความต้องการและความสนใจของลูกค้ามากที่สุด

### c. สร้างคอลัมน์ใหม่ซึ่งเป็นชุดข้อมูลใหม่ ที่เกิดจากการแปลง 4 รายการ:

````python
df_new = pd.concat(df1.values())
````

#### (1) เปลี่ยนเพศเป็นรายการ "Male" หรือ "Female"

````python
df_new[['Gender']] = df_new[['Gender']].replace([1], 'Male')
df_new[['Gender']] = df_new[['Gender']].replace([2], 'Female')
````

#### (2) เปลี่ยน Children เป็นรายการ "No children" หรือ "At least one child"

````python
df_new[['Children']] = df_new[['Children']].replace([0], 'No children')
df_new[['Children']] = df_new[['Children']].replace([1, 2, 3], 'At least one child')
````

#### (3) เปลี่ยน Salary เป็นรายการ "Less than $40K", "Between $40K and $70K", "Between $70K and $100K" และ "Greater than $100K"

````python
conditions = [
    (df_new['Salary'] < 40000),
    (df_new['Salary'] >= 40000) & (df_new['Salary'] <= 70000),
    (df_new['Salary'] > 70000) & (df_new['Salary'] <= 100000),
    (df_new['Salary'] > 100000)
]

values = ['Less than $40K', 'Between $40K and $70K', 'Between $70K and $100K', 'Greater than $100K']

df_new['Salary'] = np.select(conditions, values)
````

#### (4) เปลี่ยน Opinion เป็นรายการตัวเลข 1 ถึง 5 จาก Strongly Disagree, Disagree, Neutral, Agree และ Strongly Agree ตามลำดับ

````python
conditions1 = [
    (df_new['Opinion'] == 'Strongly disagree'),
    (df_new['Opinion'] == 'Disagree'),
    (df_new['Opinion'] == 'Neutral'),
    (df_new['Opinion'] == 'Agree'),
    (df_new['Opinion'] == 'Strongly agree'),
]

values1 = [1, 2, 3, 4, 5]

df_new['Opinion'] = np.select(conditions1, values1)
````

### d. สร้างแผนภูมิของแต่ละคอลัมน์ Salary ที่สร้างขึ้นมาใหม่

````python
df_new['Salary'].value_counts().plot(kind='bar', figsize=[5, 5], xlabel='Salary Group', ylabel='Number of Person', title='Salary bar chart', color=['red', 'green', 'blue', 'yellow'])
````
